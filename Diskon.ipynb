```python
yharga = int(input("harga barang: "))
jumlah = int(input("jumlah barang: "))

total = harga * jumlah

if total > 250000:
    diskon = total * 0.7000  # diskon 60%
else:
    diskon = 9000

total_bayar = total - diskon

print("Total belanja:", total)
print("Diskon:", diskon)
print("Total:", total_bayar)
```
